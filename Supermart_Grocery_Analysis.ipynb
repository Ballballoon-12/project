```python
Total_profit_by_YearMonth = """
    SELECT
      SUM(Profit) as Total_profit,
      month || '-'|| year as year_month
    FROM
      df_clean
    GROUP BY
      year,month
    ORDER BY
      year,month;
"""

Total_profit_by_YearMonth = sqldf(Total_profit_by_YearMonth)

Total_profit_by_YearMonth['year_month'] = pd.to_datetime(Total_profit_by_YearMonth['year_month'], format='%m-%Y')

print("\n---  SQL ---")
display(Total_profit_by_YearMonth)
```

```python
import plotly.express as px


fig = px.scatter(Total_profit_by_YearMonth,
                 x='year_month',
                 y="Total_profit",
                 title="Trends of Profit with Trend Line",
                 trendline="ols",
                 trendline_color_override="red"
                )

fig.data[0].mode = 'lines'

fig.write_image("my_chart.png")

fig.show()
```

```python
cout_region = """
    SELECT
      COUNT(DISTINCT "Order ID") as "Order Count",
      Region
    FROM
      df_clean
    GROUP BY
      Region
    ORDER BY
      "Order Count";
"""

cout_region = sqldf(cout_region)

print("\n---  SQL ---")
display(cout_region)
```

```python
total_orders = cout_region['Order Count'].sum()
cout_region['Percentage'] = (cout_region['Order Count'] / total_orders) * 100

fig = px.bar(cout_region,
             x=cout_region['Region'],
             y=cout_region['Order Count'],
             text=cout_region['Percentage'].apply(lambda x: f'{x:.2f}%'),
             title="Regional share of sales (%)")

fig.update_traces(textposition='outside')
fig.show()
```

```python
Category_Profit = """
    SELECT
      Category ,
      SUM(Profit)/(SELECT SUM(Profit) FROM df_clean) as Profit_Percentage,
      SUM(Profit) as Total_profit
    FROM
      df_clean
    GROUP BY
      Category
    ORDER BY
      SUM(Profit) desc;
"""

Category_Profit = sqldf(Category_Profit)

print("\n---  SQL ---")
display(Category_Profit)
```

```python
Category_Profit['custom_text'] = [
    f"Profit: {profit:,.0f}<br><b>{percentage:.2%}</b>"
    for profit, percentage in zip(Category_Profit['Total_profit'], Category_Profit['Profit_Percentage'])
]

fig = px.bar(Category_Profit,
             x='Category',
             y='Profit_Percentage',
             text='custom_text',
             title="Profit and Percentage by Category")
fig.update_traces(textposition='outside')
fig.update_yaxes(range=[0.1, 0.16],showticklabels=False)
fig.show()
```

```python
Profit_by_city = """
    SELECT
      City,
      State,
      SUM(Profit) as Total_profit,
      row_number() over(order by SUM(Profit) desc) as RNK
    FROM
      df_clean
    GROUP BY
      City,
      State
    ORDER BY
      SUM(Profit) desc
    LIMIT 10;
"""

Profit_by_city = sqldf(Profit_by_city)

display(Profit_by_city)
```

# RFM Analysis

```python
snapshot_date = df_clean['Order Date'].max() + pd.Timedelta(days=1)
print(f"Snapshot date: {snapshot_date}")


rfm_df = df_clean.groupby('Customer Name').agg({
    'Order Date': lambda date: (snapshot_date - date.max()).days, # Recency
    'Order ID': 'count', # Frequency
    'Sales': 'sum' # Monetary
})


rfm_df.rename(columns={'Order Date': 'Recency',
                       'Order ID': 'Frequency',
                       'Sales': 'Monetary'}, inplace=True)

print("\n--- RFM Values ---")
print(rfm_df)
```

```python
r_labels = range(5, 0, -1) # [5, 4, 3, 2, 1]
fm_labels = range(1, 6)    # [1, 2, 3, 4, 5]

rfm_df['R_Score'] = pd.qcut(rfm_df['Recency'], q=5, labels=r_labels, duplicates='drop')
rfm_df['F_Score'] = pd.qcut(rfm_df['Frequency'].rank(method='first'), q=5, labels=fm_labels, duplicates='drop')
rfm_df['M_Score'] = pd.qcut(rfm_df['Monetary'], q=5, labels=fm_labels, duplicates='drop')

print("\n--- RFM Scores ---")
print(rfm_df)
```

```python
rfm_df['RFM_Score'] = rfm_df['R_Score'].astype(str) + rfm_df['F_Score'].astype(str) + rfm_df['M_Score'].astype(str)


rfm_df['RFM_Score'] = rfm_df['R_Score'].astype(str) + rfm_df['F_Score'].astype(str) + rfm_df['M_Score'].astype(str)


segt_map = {
    r'[4-5][4-5][4-5]': 'Champions',
    r'1[4-5][4-5]': 'Cant Lose Them',
    r'[2-4][3-4][4-5]': 'Loyal Customers',
    r'[1-2][2-5][2-5]': 'At Risk',
    r'[3-4][3-4][3-4]': 'Need Attention',
    r'[3-5][1-3][1-3]': 'Potential Loyalist',
    r'[4-5]11': 'New Customers',
    r'[3-4]11': 'Promising',
    r'[2-3][1-2][1-2]': 'About To Sleep',
    r'[2-3][2-3][2-3]': 'Hibernating',
    r'111': 'Lost',
}


rfm_df['Segment'] = rfm_df['RFM_Score'].replace(segt_map, regex=True)

rfm_df.loc[rfm_df['Segment'].str.match(r'^\d{3}$'), 'Segment'] = 'Others'


print("\n--- Final RFM Segmentation (Regex Method) ---")
print(rfm_df[['Recency', 'Frequency', 'Monetary', 'RFM_Score', 'Segment']].sort_values(by='RFM_Score', ascending=False))
```

```python
rfm_df.reset_index(inplace=True)
segment_counts = rfm_df['Segment'].value_counts()
segment_percentage = (segment_counts / segment_counts.sum()) * 100

fig = px.bar(x=segment_percentage.index,
             y=segment_percentage.values,
             text=segment_percentage.apply(lambda x: f'{x:.2f}%'),
             title="Customer Segmentation by Percentage")

fig.update_layout(xaxis_title="Segment", yaxis_title="Percentage (%)")
fig.update_traces(textposition='outside')
fig.show()
```

```python
# Calculate the total monetary value for each segment
segment_monetary = rfm_df.groupby('Segment')['Monetary'].sum().reset_index()

# Create the treemap
fig = px.treemap(segment_monetary,
                 path=['Segment'],
                 values='Monetary',
                 title='Customer Segmentation by Monetary Value')

fig.show()
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

basket = (df_clean.groupby(['Order ID', 'Sub Category'])['Sub Category']
          .count().unstack().reset_index().fillna(0)
          .set_index('Order ID'))

# 2.  One-Hot Encoding (0  1)
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

basket_encoded = basket.applymap(encode_units)

print("---  ---")
display(basket_encoded)
```

```python
#  apriori,  0/1  False/True
basket_bool = basket_encoded.astype('bool')

#  DataFrame  bool 
frequent_itemsets = apriori(basket_bool, min_support=0.01, use_colnames=True)

print(frequent_itemsets)
```

```python
#  "confidence"    0.5 (50%)
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)

# 
rules = rules.sort_values('confidence', ascending=False)

print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])
```
